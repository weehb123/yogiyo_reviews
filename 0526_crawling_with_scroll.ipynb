```python
import os, re
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import urllib.request as ur
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

browser =  webdriver.Chrome('./chromedriver.exe')

review_cate = ['taste', 'service', 'mood', 'price', 'amount', 'menu', 'cleanliness', 'waitingtime', 'total', 'location']

taste_list = []
service_list = []
mood_list = []
price_list = []
amount_list = []
menu_list = []
cleanliness_list = []
waitingtime_list = []
total_list = []
location_list = []
cate_list = [taste_list, service_list,mood_list,price_list,amount_list,menu_list,cleanliness_list,waitingtime_list,total_list,location_list]


for i, cate in enumerate(review_cate):
    print(cate, 'keywords')
    
    #주소 가져오기
    browser.get("https://m.place.naver.com/restaurant/36096945/review/visitor?reviewTheme={0}".format(cate))
    browser.maximize_window()


    # 더보기 클릭
    try :
        more = browser.find_element_by_css_selector('div._2kAri a._3iTUo')
        more.click()
    except :
        pass

    # 화면 가장 아래로 스크롤 내리기
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    Interval = 3 # 2초에 한번씩

    # 현재 문서 높이를 가져와서 저장
    prev_height = browser.execute_script("return document.body.scrollHeight")

    while True:
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        try:
            more.click()
        except:
            pass
        time.sleep(Interval)
    
        curr_height = browser.execute_script("return document.body.scrollHeight")
        if curr_height == prev_height:
            break
    
        prev_height = curr_height

    print("스크롤 완료")

    # 하이라이트 키워드 스크랩

    elems = browser.find_elements_by_css_selector('span.highlight')
    for elem in elems:
        cate_list[i].append(elem.text)

        
print(cate_list)

        
```
